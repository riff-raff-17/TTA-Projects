```python
import pygame
import random

# Initialize Pygame
pygame.init()

# Screen dimensions
WIDTH, HEIGHT = 800, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Top-Down Shooter - Sprites Version")

# Load assets
player_img = pygame.image.load("player.png")
player_img = pygame.transform.scale(player_img, (50, 50))
bullet_img = pygame.image.load("bullet.png")
bullet_img = pygame.transform.scale(bullet_img, (10, 20))
enemy_img = pygame.image.load("enemy.png")
enemy_img = pygame.transform.scale(enemy_img, (50, 50))

# Player settings
player_pos = [WIDTH // 2, HEIGHT - 60]
player_speed = 5

# Bullet settings
bullets = []
bullet_speed = 7

# Enemy settings
enemies = []
enemy_speed = 2

# Game clock
clock = pygame.time.Clock()

# Spawn enemies
def spawn_enemy():
    x = random.randint(0, WIDTH - 50)
    y = random.randint(-100, -40)
    enemies.append([x, y])

# Main game loop
running = True
while running:
    screen.fill((255, 255, 255))
    
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_SPACE:
                # Shoot bullet
                bullets.append([player_pos[0] + 20, player_pos[1]])

    # Player movement
    keys = pygame.key.get_pressed()
    if keys[pygame.K_LEFT] and player_pos[0] > 0:
        player_pos[0] -= player_speed
    if keys[pygame.K_RIGHT] and player_pos[0] < WIDTH - 50:
        player_pos[0] += player_speed

    # Update bullets
    for bullet in bullets[:]:
        bullet[1] -= bullet_speed
        if bullet[1] < 0:
            bullets.remove(bullet)

    # Spawn and update enemies
    if random.randint(1, 30) == 1:  # Randomly spawn enemies
        spawn_enemy()
    
    for enemy in enemies[:]:
        enemy[1] += enemy_speed
        if enemy[1] > HEIGHT:
            enemies.remove(enemy)  # Remove enemy if it goes off-screen

        # Check for collision with bullets
        for bullet in bullets[:]:
            if pygame.Rect(enemy[0], enemy[1], 50, 50).colliderect(
                pygame.Rect(bullet[0], bullet[1], 10, 20)
            ):
                bullets.remove(bullet)
                enemies.remove(enemy)
                break

    # Draw player
    screen.blit(player_img, player_pos)

    # Draw bullets
    for bullet in bullets:
        screen.blit(bullet_img, bullet)

    # Draw enemies
    for enemy in enemies:
        screen.blit(enemy_img, enemy)

    # Update display
    pygame.display.flip()
    clock.tick(60)

quit()
```
