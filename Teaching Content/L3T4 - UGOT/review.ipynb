## 2. Data Types and Variables

Common data types:
- `int` (whole numbers): `5`, `-2`
- `float` (decimals): `3.14`, `-0.5`
- `str` (text): `"hello"`
- `bool` (True/False): `True`, `False`

Variables store values:
```python
age = 16
name = "Sam"
```

You can check the type with `type`:

```python
type(age) = int
type(name) = string
```

````python
age = input("Enter your age: ")
print("Age:", age)
print("Type of age:", type(age))
````

## 3. Operators

### Arithmetic
- `+` add
- `-` subtract
- `*` multiply
- `/` divide (always float)
- `//` integer division
- `%` remainder (mod)
- `**` exponent

### Comparison (results in bool)
- `==`, `!=`, `<`, `<=`, `>`, `>=`

### Logical
- `and`, `or`, `not`


````python
print("7 / 2 =", 7 / 2)
print("7 // 2 =", 7 // 2)
print("7 % 2 =", 7 % 2)
print("2 ** 5 =", 2 ** 5)

print("3 == 3:", 3 == 3)
print("3 != 4:", 3 != 4)
print("(5 > 2) and (1 > 3):", (5 > 2) and (1 > 3))
````

## 4. if / elif / else

Conditional logic lets your program make decisions.

```python
if condition:
    ...
elif another_condition:
    ...
else:
    ...
```


````python
temp = float(input("Enter the temperature in °C: "))

if temp < 0:
    print("Freezing!")
elif temp < 20:
    print("Cool.")
elif temp < 30:
    print("Warm.")
else:
    print("Hot!")
````

## 5. Loops

Loops repeat code.

### `while` loop
Repeats *while* a condition is True.

### `for` loop
Repeats over a sequence (often `range(...)` or a list).


````python
count = 1
while count <= 5:
    print("Count:", count)
    count += 1
````

````python
for i in range(1, 6):
    print("i =", i)
````

## 6. Functions

Functions help you:
- reuse code
- organize logic
- make programs easier to read and test

```python
def name(parameters):
    # body
    return value  # optional


````python
def greet(name):
    return f"Hello, {name}!"

message = greet("Aisha")
print(message)
````

````python
def area_rectangle(width, height):
    return width * height

print(area_rectangle(3, 4))
````
