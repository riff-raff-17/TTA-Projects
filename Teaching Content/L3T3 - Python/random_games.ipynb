```python
"""Number guessing game"""

import random

number = random.randint(1, 100)
attempts = 0
guess = None

print("I'm thinking of a number between 1 and 100.")

while guess != number:
    guess = int(input("Your guess: "))
    attempts += 1
    if guess < number:
        print("Too low!")
    elif guess > number:
        print("Too high!")
    else:
        print(f"Correct! You guessed it in {attempts} tries.")
```

```python
"""Magic 8 ball"""

import random

answers = [
    "Yes, definitely.",
    "No way.",
    "Ask again later.",
    "Absolutely!",
    "I wouldn't count on it.",
    "It's uncertain..."
]

while True:
    question = input("Ask the Magic 8 Ball a question (or type 'quit'): ")
    if question.lower() == 'quit':
        break
    print(random.choice(answers))
```

```python
"""RPS"""

import random

# List of possible moves
options = ["rock", "paper", "scissors"]

computer = options[random.randint(0, 2)]

player = input("Choose rock, paper, or scissors: ").strip().lower()

print("You chose:", player)
print("Computer chose:", computer)

# Who won?
if player == computer:
    print("It is a draw ._.")

elif (player == 'paper' and computer == "rock") or \
    (player == 'scissors' and computer == "paper") or \
    (player == 'rock' and computer == "scissors"):
    print("You won")

else:
    print("You lose")
```

```python
'''Simple Hangman'''
import random

# 1) Words to choose from
words = [
    "python", "random", "hangman", "variable", "string",
    "loop", "function", "integer", "boolean", "module"
]

# 2) Pick a secret word
secret = random.choice(words)

# 3) Game state
lives = 6 # how many wrong guesses allowed
guessed_letters = "" # letters the player has tried
display = ["_"] * len(secret) # current view of the word

print("=== Hangman ===")
print("I'm thinking of a word with", len(secret), "letters.")

# 4) Main game loop
while True:
    print("\nWord: ", " ".join(display))
    print("Lives left:", lives)
    print("Guessed so far:", " ".join(guessed_letters))

    guess = input("Guess a single letter: ").lower().strip()

    # Basic input checks
    if guess == 'quit':
        print("Bye!")
        break
    if len(guess) != 1 or not guess.isalpha():
        print("Please type exactly ONE letter (a-z).")
        continue
    if guess in guessed_letters:
        print("You already guessed that letter.")
        continue

    guessed_letters += guess  # remember this guess

    if guess in secret:
        # Reveal all positions of that letter
        for i in range(len(secret)):
            if secret[i] == guess:
                display[i] = guess
        print("Nice! That letter is in the word.")
    else:
        lives -= 1
        print("Nope! That letter is not in the word.")

    # Win condition: no underscores left
    if "_" not in display:
        print("\nYou win! The word was:", secret)
        break

    # Lose condition: no lives left
    if lives == 0:
        print("\nOut of lives! The word was:", secret)
        break
```

```python
'''Simple Hangman with clear_output'''
from IPython.display import clear_output
import random

# 1) Words to choose from
words = [
    "python", "random", "hangman", "variable", "string",
    "loop", "function", "integer", "boolean", "module"
]

# 2) Pick a secret word
secret = "brain"

# 3) Game state
lives = 6
guessed_letters = ""
display = ["_"] * len(secret)

# 4) Main game loop
while True:
    # Print updated board
    clear_output(wait=True)
    print("=== Hangman ===")
    print("\nWord: ", " ".join(display))
    print("Lives left:", lives)
    print("Guessed so far:", " ".join(sorted(guessed_letters)) if guessed_letters else "(none)")

    guess = input("Guess a single letter: ").lower().strip()

    # Basic input checks
    if guess == 'quit':
        print("Bye!")
        break
    if len(guess) != 1 or not guess.isalpha():
        print("Please type exactly ONE letter (a-z).")
        input("Press Enter to continue...")
        continue
    if guess in guessed_letters:
        print("You already guessed that letter.")
        input("Press Enter to continue...")
        continue

    # Clear only after a valid guess
    clear_output(wait=True)
    guessed_letters += guess  # remember this guess

    if guess in secret:
        for i in range(len(secret)):
            if secret[i] == guess:
                display[i] = guess
    else:
        lives -= 1

    # Win condition
    if "_" not in display:
        clear_output(wait=True)
        print("You win! The word was:", secret)
        break

    # Lose condition
    if lives == 0:
        print("\nOut of lives! The word was:", secret)
        break
```

```python
'''Complex Hangman'''

import random
import string

HANGMANPICS = [
    """
     +---+
     |   |
         |
         |
         |
         |
    =========""",
    """
     +---+
     |   |
     O   |
         |
         |
         |
    =========""",
    """
     +---+
     |   |
     O   |
     |   |
         |
         |
    =========""",
    """
     +---+
     |   |
     O   |
    /|   |
         |
         |
    =========""",
    """
     +---+
     |   |
     O   |
    /|\\  |
         |
         |
    =========""",
    """
     +---+
     |   |
     O   |
    /|\\  |
    /    |
         |
    =========""",
    """
     +---+
     |   |
     O   |
    /|\\  |
    / \\  |
         |
    ========="""
]

WORDS = [
    "python", "variable", "function", "random", "hangman",
    "notebook", "iterator", "boolean", "package", "integer",
    "recursion", "class", "object", "string", "exception",
    "module", "argument", "keyword", "operator", "dictionary"
]

def choose_secret(words):
    return random.choice(words)

def display_state(secret, guessed_letters, wrong_guesses):
    print(HANGMANPICS[len(wrong_guesses)])
    display = " ".join([c if c in guessed_letters else "_" for c in secret])
    print(f"\nWord: {display}")
    if wrong_guesses:
        print("Wrong:", " ".join(sorted(wrong_guesses)))
    remaining = len(HANGMANPICS) - 1 - len(wrong_guesses)
    print(f"Lives left: {remaining}\n")

def get_guess(already_guessed):
    while True:
        guess = input("Guess a letter: ").strip().lower()
        if len(guess) != 1 or guess not in string.ascii_lowercase:
            print("Please enter a single letter (a-z).")
            continue
        if guess in already_guessed:
            print("You already guessed that letter. Try a different one.")
            continue
        return guess

def play_round():
    secret = choose_secret(WORDS)
    guessed_letters = set()
    wrong_guesses = set()
    max_mistakes = len(HANGMANPICS) - 1

    while True:
        display_state(secret, guessed_letters, wrong_guesses)

        if all(c in guessed_letters for c in secret):
            print(f"You won! The word was: {secret}")
            break

        if len(wrong_guesses) >= max_mistakes:
            print(f"Game over! The word was: {secret}")
            break

        guess = get_guess(guessed_letters | wrong_guesses)

        if guess in secret:
            print(f"Nice! '{guess}' is in the word.\n")
            guessed_letters.add(guess)
        else:
            print(f"Oops‚Äîno '{guess}'.\n")
            wrong_guesses.add(guess)

def main():
    print("=== Hangman (random-powered) ===")
    while True:
        play_round()
        again = input("\nPlay again? (y/n): ").strip().lower()
        if again != "y":
            print("Thanks for playing!")
            break
        print("\n" + "=" * 28)

main()
```

```python
import random

print("üéâ Welcome to Confetti Pop Roulette!")
print("Two players take turns popping party poppers.")
print("Confetti (üéä) costs a ‚≠ê. Duds (üß®) do nothing.\n")

# Random number of confetti poppers (1‚Äì6), rest are duds
num_confetti = random.randint(1, 6)
num_duds = 6 - num_confetti

# Build and shuffle the popper lineup (1 = confetti, 0 = dud)
poppers = [1] * num_confetti + [0] * num_duds
random.shuffle(poppers)

# Player stars (health)
p1_stars = 3
p2_stars = 3
current_player = 1  # 1 or 2

print(f"üß∫ Basket loaded ‚Äî CONFETTI: {num_confetti}, DUDS: {num_duds} (total {len(poppers)})")
print("Type 'me' to pop your own, or 'opponent' to pop the other player's.\n")

def show_status():
    print(f"‚≠ê Player 1: {p1_stars}   ‚≠ê Player 2: {p2_stars}")

while poppers and p1_stars > 0 and p2_stars > 0:
    print("\n------------------------------------")
    print(f"üéØ Player {current_player}'s turn")
    confetti_left = poppers.count(1)
    duds_left = poppers.count(0)
    print(f"üìä Remaining ‚Äî CONFETTI: {confetti_left}, DUDS: {duds_left}")
    show_status()

    choice = input("Pop 'me' or 'opponent'? ").strip().lower()
    if choice not in {"me", "opponent"}:
        print("Please type 'me' or 'opponent'.")
        continue

    # Draw the next popper
    pop = poppers.pop(0)

    # Who gets popped?
    target_is_p1 = (current_player == 1 and choice == "me") or (current_player == 2 and choice == "opponent")

    if pop == 1:
        print("üéä CONFETTI! Sparkles everywhere!")
        if target_is_p1:
            p1_stars -= 1
        else:
            p2_stars -= 1
    else:
        print("üß® Dud. Nothing happens.")

    # Check for game end
    if p1_stars <= 0:
        print("\nüèÅ Player 1 is out of stars. Player 2 wins! üèÜ")
        break
    if p2_stars <= 0:
        print("\nüèÅ Player 2 is out of stars. Player 1 wins! üèÜ")
        break

    # If it was a dud, turn passes; if confetti, same player keeps the turn (feels rewarding!)
    if pop == 0:
        current_player = 2 if current_player == 1 else 1

# If all poppers are used and both still have stars
if not poppers and p1_stars > 0 and p2_stars > 0:
    print("\nüí® All poppers used! Final stars:")
    show_status()
    if p1_stars > p2_stars:
        print("üèÜ Player 1 wins on stars!")
    elif p2_stars > p1_stars:
        print("üèÜ Player 2 wins on stars!")
    else:
        print("ü§ù It's a cheerful tie!")
```
